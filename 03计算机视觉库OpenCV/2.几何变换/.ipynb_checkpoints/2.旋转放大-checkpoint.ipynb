```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
    
def imread(image):
    image = cv2.imread(image)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    return image
```

```python
image = imread('test.jpg')
show(image)
```

```python
image = imread('test.jpg')
(h,w) = image.shape[:2]
(cX,cY) = (w/2, h/2)

# 逆时针旋转45度
M = cv2.getRotationMatrix2D((cX, cY), 45, 1.0)
# 输入图像，M，输出图像的大小
image = cv2.warpAffine(image, M, (w,h))
show(image)
```

```python
image = imread('test.jpg')
(h,w) = image.shape[:2]
(cX,cY) = (w/2, h/2)

# 顺时针旋转90度
M = cv2.getRotationMatrix2D((cX, cY), -90, 1.0)
# 输入图像，M，输出图像的大小
image = cv2.warpAffine(image, M, (w,h))
show(image)
```

```python
image = imread('test.jpg')
(h,w) = image.shape[:2]
(cX,cY) = (w/2, h/2)

# 逆时针旋转45度
M = cv2.getRotationMatrix2D((cX-100, cY-100), 45, 1.0)
# 输入图像，M，输出图像的大小
image = cv2.warpAffine(image, M, (w,h))
show(image)
```

```python
image = imread('test.jpg')
(h,w) = image.shape[:2]
(cX,cY) = (w/2, h/2)

# 逆时针旋转45度，并放大2倍
M = cv2.getRotationMatrix2D((cX, cY), 45, 2.0)
# 输入图像，M，输出图像的大小
image = cv2.warpAffine(image, M, (w,h))
show(image)
```
