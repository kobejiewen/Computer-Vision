```python
import cv2
import matplotlib.pyplot as plt
```

```python
# 读入图片
image = cv2.imread('test.jpg')
# 高度宽度通道数
(h, w, c) = image.shape
image.shape
```

```python
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
plt.axis('off')
plt.show()
```

```python
# 通常的颜色是RGB，OpenCV保存的颜色是BGR()
# 图片中一个点的RGB三颜色的值
(b, g, r) = rgb_image[0, 0]
print("Pixel at (0,0):Red:{r}, Green:{g}, Blue:{b}".format(r=r,g=g,b=b))
```

```python
# 改变一个点的值
rgb_image[0, 0] = (0, 0, 255)
(b, g, r) = image[0, 0]
print("Pixel at (0,0):Red:{r}, Green:{g}, Blue:{b}".format(r=r,g=g,b=b))
```

```python
# 图像中点
cX, cY = (w//2 , h//2)
# 左上角的部分
tl = rgb_image[0:cY,0:cX]
plt.imshow(tl)
plt.axis('off')
plt.show()
```

```python
# 右上方
tr = rgb_image[0:cY, cX:w]
# 左下方
bl = rgb_image[cY:h, 0:cX]
# 右下方
br = rgb_image[cY:h, cX:w]

# 定义一个函数用于显示图片
def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()

# plt.imshow(tl)
# plt.axis('off')
# plt.show()
# plt.imshow(tr)
# plt.axis('off')
# plt.show()
# plt.imshow(bl)
# plt.axis('off')
# plt.show()
# plt.imshow(br)
# plt.axis('off')
# plt.show()
show(tl)
show(tr)
show(bl)
show(br)
```

```python
rgb_image[0:cY,0:cX] = (0, 255, 0)

show(rgb_image)
```
